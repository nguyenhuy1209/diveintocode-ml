```
"""
Problem 1
Creating random numbers
"""

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

sample = np.random.multivariate_normal(mean=(-3,0), cov=[[1.0, 0.8], [0.8, 1.0]], size=(500))

print(f'First 10 samples of sample')
print(sample[0:10])
print(f'Size of sample {sample.shape}')
```

```
"""
Problem 2
Visualization by scatter plot
"""

plt.scatter(sample[:,0], sample[:,1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.show()
```

```
"""
Problem 3
Visualization by histogram
"""

plt.hist(sample[:,0], 50)
plt.xlim(-5.55, 2.7)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")
plt.show()

plt.hist(sample[:,1], 50)
plt.xlim(-5.55, 2.7)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x2")
plt.show()
```

```
"""
Problem 4
Addition of data
"""

sample2 = np.random.multivariate_normal(mean=(0,-3), cov=[[1.0, 0.8], [0.8, 1.0]], size=(500))

plt.scatter(sample[:,0], sample[:,1], label="0")
plt.scatter(sample2[:,0], sample2[:,1], label="1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.legend(loc="upper right")
plt.show()
```

```
"""
Problem 5
Data combination
"""

merged_sample = np.concatenate((sample,sample2))

print(f'First 10 samples of merged_sample')
print(merged_sample[0:10])
print(f'Size of merged sample {merged_sample.shape}')
```

```
"""
Problem 6
Labeling
"""
addition_vector = np.array([0 if i < 500 else 1 for i in range(1000)])
addition_vector = np.reshape(addition_vector, (1000,1))

labelled_sample = np.hstack((merged_sample,addition_vector))

print(f'First 10 samples of labelled_sample')
print(labelled_sample[0:10])
print(f'Size of labelled sample {labelled_sample.shape}')
```

**Problem 1**

Data acquisition

```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
Y = pd.DataFrame(iris.target)

print(X)
print(Y)
```

**Problem 2**

Combining data

```python
df = pd.concat([X, Y], axis=1)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Species']
print(df)
```

**Problem 4** <br>

**Context** <br>
The Iris flower data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

This dataset became a typical test case for many statistical classification techniques in machine learning such as support vector machines

**Content** <br>
The dataset contains a set of 150 records under 5 attributes - Petal Length, Petal Width, Sepal Length, Sepal width and Class(Species).

**Problem 6**

Creating a diagram

```python
"""
Problem 6
"""

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = df.groupby("Species")['sepal_length'].count()

labels = iris.target_names
plt.pie(x=data, autopct="%.1f%%", labels=labels, pctdistance=0.5)
plt.title("Number of samples per label", fontsize=14)
plt.show()
```

```python
"""
Select features one by one and visualize the distribution of data for each label using a box plot
"""

print('Select features one by one and visualize the distribution of data for each label using a box plot')
def drawBoxPlot(data, title):
    fig, axs = plt.subplots(2, 2)
    axs[0, 0].set_title('Sepal length')
    axs[0, 0].boxplot(data.iloc[:, 0])

    axs[0, 1].set_title('Sepal width')
    axs[0, 1].boxplot(data.iloc[:, 1])

    axs[1, 0].set_title('Petal length')
    axs[1, 0].boxplot(data.iloc[:, 2])

    axs[1, 1].set_title('Petal width')
    axs[1, 1].boxplot(data.iloc[:, 3])

    plt.tight_layout()
    fig.suptitle(title)
    plt.show()

setosa = df.loc[df['Species'] == 0]
versicolor = df.loc[df['Species'] == 1]
virginica = df.loc[df['Species'] == 2]

drawBoxPlot(setosa, 'Setosa')
drawBoxPlot(versicolor, 'Versicolor')
drawBoxPlot(virginica, 'Virginica')
```

```python
"""
Select features one by one and visualize the distribution of data for each label using a violin plot
"""

print('Select features one by one and visualize the distribution of data for each label using a violin plot')
def drawViolinPlot(data, title):
    fig, axs = plt.subplots(2, 2)

    axs[0,0].set_title('Sepal length')
    axs[0,0].violinplot(data.iloc[:, 0])

    axs[0,1].set_title('Sepal width')
    axs[0,1].violinplot(data.iloc[:, 1])

    axs[1,0].set_title('Petal length')
    axs[1,0].violinplot(data.iloc[:, 2])

    axs[1,1].set_title('Petal width')
    axs[1,1].violinplot(data.iloc[:, 3])

    plt.tight_layout()
    fig.suptitle(title)
    plt.show()

drawViolinPlot(setosa, 'Setosa')
drawViolinPlot(versicolor, 'Versicolor')
drawViolinPlot(virginica, 'Virginica')
```

A **violin plot** is more informative than a plain **box plot**. 

While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

**Problem 7**

Confirming the relationship between features

```python
"""
Take one feature for each of the vertical and horizontal axes and create a color-coded scatter plot for each type
"""

def drawScatter(data, title):
    fig, axs = plt.subplots(1, 2)

    petal_length = data.loc[:, 'petal_length']
    petal_width = data.loc[:, 'petal_width']
    sepal_length = data.loc[:, 'sepal_length']
    sepal_width = data.loc[:, 'sepal_width']

    axs[0].set_title('Petal')
    axs[0].scatter(petal_length, petal_width)
    axs[0].set(xlabel='Length', ylabel='Width')

    axs[1].set_title('Sepal')
    axs[1].scatter(sepal_length, sepal_width)
    axs[1].set(xlabel='Length', ylabel='Width')

    plt.tight_layout()
    fig.suptitle(title)
    plt.show()

# Take one feature for each of the vertical and horizontal axes and create a color-coded scatter plot for each type (6).
drawScatter(setosa, 'Setosa')
drawScatter(versicolor, 'Versicolor')
drawScatter(virginica, 'Virginica')
```

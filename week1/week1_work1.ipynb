```
"""
Problem 1
Create using the exponentiation arithmetic operator
"""
THICKNESS = 0.00008
folded_thickness_43_times = THICKNESS*2**43
print("Thickness: {} meters".format(folded_thickness_43_times))
```

```
"""
Problem 2
Unit Conversion
"""
# Convert meters to 10000 kilometers and display with two decimal places
print("Thickness: {: .2f} x 10000 kilometers".format(folded_thickness_43_times/(1000*10000)))
# The distance to the moon is 384.400 km, so folding a paper 43 times leads you far beyond the moon
```

```
"""
Problem 3
Create using a for statement
"""
folded_thickness_43_times = THICKNESS
for i in range(43):
  folded_thickness_43_times *= 2
print("Thickness: {} meters".format(folded_thickness_43_times))
```

```
"""
Problem 4
Create using a for statement
"""
import time

THICKNESS = 0.00008

# 
start = time.time()
folded_thickness = THICKNESS*2**43
elapsed_time = time.time() - start
print("Thickness: {} meters".format(folded_thickness))
print("time : {}[s]".format(elapsed_time))

# 
start = time.time()
folded_thickness_43_times = THICKNESS
for i in range(43):
  folded_thickness_43_times *= 2
elapsed_time = time.time() - start
print("Thickness: {} meters".format(folded_thickness_43_times))
print("time : {}[s]".format(elapsed_time))
```

```
"""
Problem 5
Saving to a list
"""

l = [THICKNESS]

for i in range(43):
  l.append(l[-1] * 2)

print(len(l))
```

```
"""
Problem 6
Displaying a line graph
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(l)
plt.show()

# The thickness grows exponentially with the growth rate = 2
```

```
"""
Problem 1
Number of grains of rice on the 100th day
"""
import matplotlib.pyplot as plt

RICE = 1
DAYS = 100

l = [RICE]

for i in range(DAYS):
  l.append(l[-1] * 2)

# Number of rice on the 100th day
print(f'Number of rice on the 100th day:', l[DAYS])

# Plot the line graph
%matplotlib inline
plt.title("Sorori Shinzaemon problem")
plt.xlabel("Number of days")
plt.ylabel("Number of rice")
plt.plot(l)
plt.show()
```

```
"""
Problem 2
Number of rice grains outside of the 100th day
"""

import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(day=100):
  """
  A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
  Parameteres
  --------------
  day: int
      Number of days to get rice (default : 100)
  Returns
  --------------
  list_n_grains : list
      List of the number of rice you get one day (the length of the list is equal to day)
  list_total_grains : list
      List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
  """
  # Write the code here. pass is an instruction that means do nothing.
  import matplotlib.pyplot as plt

  RICE = 1

  list_n_grains = [RICE] * day
  list_total_grains = [RICE] * day

  for i in range(1,day):
    num_of_grain_today = 2**i
    list_total_grains[i] = list_total_grains[i-1] + num_of_grain_today
    list_n_grains[i] = num_of_grain_today
  return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=20)

# Plot the line graph
%matplotlib inline
plt.title("Number of rice received in any day")
plt.xlabel("Number of days")
plt.ylabel("Number of rice")
plt.plot(list_n_grains)
plt.show()

# Plot the line graph
%matplotlib inline
plt.title("Total number of rice in a certain day")
plt.xlabel("Number of days")
plt.ylabel("Total rice")
plt.plot(list_total_grains)
plt.show()
```

```
"""
Problem 3
How many people can live for how many days with the rice we get?
"""

"""
https://en.wikipedia.org/wiki/Freerice#:~:text=The%20United%20Nations'%20World%20Food,one%20adult%20for%20a%20day.

According to this wiki, "the United Nations stated that it takes 400 grams or about 19,200 grains of rice to feed one adult for a day.",
so I would come up with this formula: 

number of day 1 person can live = number of grains / 19200

number of day n person can live = number of grains / (19200*n)

"""


def compute_day_can_live_with_rice(num_rice_grain, num_of_people):
  return num_rice_grain/(19200*num_of_people)

# number of day 1000 people can live after 30 days receiving grains
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=30)
print(f'Number of days 1000 people can live after 30 days receiving grains is: {compute_day_can_live_with_rice(list_n_grains[-1], 1000)} days')
```
